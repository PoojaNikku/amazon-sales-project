### **1. Importing Libraries** ###

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("seaborn-v0_8")
```

### **2. Loading the Dataset** ###

```python
df = pd.read_csv("amazon.csv")
df.head()
```

### **3. Initial Data Inspection** ##

```python
df.info()
df.head()
```

### **4. Data Cleaning** ###

##### **1 - Clean discount_percentage column** #####

```python
df['discount_percentage'] = df['discount_percentage'].str.replace('%', '').astype(float)
```

##### **3 ‚Äî Fix rating_count missing values** #####

```python
df['rating_count'].fillna(df['rating_count'].median(), inplace=True)
```

##### **4 ‚Äî Clean category column** #####

```python
df['main_category'] = df['category'].str.split('|').str[0]
df['sub_category'] = df['category'].str.split('|').str[1]
df['sub_sub_category'] = df['category'].str.split('|').str[2]
```

##### **5 ‚Äî Feature engineering** #####

```python
#Calculate discount amount
df['discount_amount'] = df['actual_price'] - df['discounted_price']
```

```python
#Calculate discount percentage again
df['calculated_discount_pct'] = (df['discount_amount'] / df['actual_price']) * 100
```

```python
#Calculate profit loss
df['price_difference'] = df['actual_price'] - df['discounted_price']
```

```python
#Clean category names
df['main_category'] = df['main_category'].str.replace('&', ' & ')
df['sub_category'] = df['sub_category'].str.replace('&', ' & ')
df['sub_sub_category'] = df['sub_sub_category'].str.replace('&', ' & ')
```

##### **6 ‚Äî Clean Product Names** #####

```python
df['product_name'] = df['product_name'].str.strip()
```

```python
df['product_name'] = df['product_name'].str.replace('\n', ' ')
```

##### **7 ‚Äî Remove Duplicates** #####

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

#### Discount Insights
- Electronics and Computer Accessories receive the highest average discounts (50%+).
- No strong negative correlation found between high discounts and product ratings ‚Äî discounts do not reduce product quality perception.

#### Top Products

### Recommendations

####  1. Improve Stock & Visibility for High-Demand Categories
Electronics and Computer Accessories dominate the product listings and drive engagement. Amazon should ensure continuous availability and highlight these in promotions.

####  2. Continue Aggressive Discounting in Tech Categories
Discounts >50% in Electronics and Accessories do not negatively affect ratings. Amazon can continue using discounts as a volume-growth strategy.

####  3. Boost Promotion for Premium Electronics
Premium TVs and ACs have excellent ratings (4.5+) and high engagement.
These should be promoted during:
- Amazon Great Indian Festival
- Lightning deals
- Banner ads

####  4. Expand AmazonBasics Product Line
AmazonBasics consistently ranks among the most-reviewed products with strong customer trust.
Introduce more AmazonBasics items in:
- Mobile accessories  
- Kitchen appliances  
- Home gadgets  

####  5. Improve Quality of Low-Rated Low-Price Items
Products under ‚Çπ1,000 with ratings <4.0 need:
- Quality inspection  
- SKU review  
- Customer complaint pattern analysis  

####  6. Use BI Dashboard KPIs for Weekly Monitoring
Category managers should track:
- Avg rating  
- Avg discount %  
- Review spikes or drops  
- Top 10 trending products  
- Category-wise pricing trends  

This ensures better decision-making and faster intervention on declining products.
